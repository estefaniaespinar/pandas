```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/andressalomferrer/Desktop/pokemon_data.csv/pokemon_data.csv')
```

```python
#filtrat por una coldicion de row dentro de una columna especifica
df.loc[df['Type 1'] == 'Grass'][0:3]
```

```python
#filtrar por mas de una columna con condicion:
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')][0:3]

## and en panda es & 
## or en pandas es |

#df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')][0:3]
```

```python
#Tb podemos filtrar por tres o más conditions

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
```

```python
## Filtrar por rows que contengan en la row la palaba 'mega' or 'Mega'

df.loc[(df['Name'].str.contains('Mega')) | (df['Name'].str.contains('mega'))] [0:3]
```

```python
#También podemos usar regex
import re
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)][0:3]
```

```python
#para que no nos distinga entre mayusculas y minusculas
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)][0:3]
```

```python
#Todas las rows de una columna que empiecen por 'pi'
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)][0:3]
```
