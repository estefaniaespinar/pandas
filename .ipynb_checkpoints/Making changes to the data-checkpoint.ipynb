```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/andressalomferrer/Desktop/pokemon_data.csv/pokemon_data.csv')
```

```python
df.head()
```

```python
#a√±adir una nueva columna

df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
```

```python
df.head(3)
```

```python
#how to drop a specific column

df.drop('Total', axis=1)[0:3]
```

```python
#otra forma m√°s f√°cil de a√±adir una columna # 4 pq empieza en HP pero 10 pq el ultimo
#no est√° incluido as√≠ que para que incluya Speed, hemos de poner 10 en vez de 9
#axis 1 para que sume las rows

df['Total'] = df.iloc[:,4:10].sum(axis=1)
```

```python
df.head(3)
```

```python
#Cambiar el orden de las columnas

df = df[['Name', 'Type 1','Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total', 'Generation', 'Legendary']]
```

```python
df.head(3)
```

```python
##volvemos a poner el Total a lo ultimo para ver como podemos hacerlo de otra forma

df = df[['Name', 'Type 1','Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary', 'Total']]
```

```python
df.head(3)
```

```python
##Tambi√©n puedes cambiar el orden de las columnas de una manera m√°s rapida:

cols = list(df.columns.values)
cols
```

```python
#cols [-1] va entre brackets pq cuando pones un solo n√∫mero lo reconoce como string
# y lo tenemos que pasar a lista ((-1 es para que nos coja el ultimo valor))

df = df[cols[0:4]+[cols[-1]]+cols[4:11]]
```
